```python
from pypdf import PdfReader, PdfWriter

def extract_pages(input_pdf, output_pdf, start_page, end_page):
    reader = PdfReader(input_pdf)
    writer = PdfWriter()

    # We subtract 1 from start_page because pypdf uses 0-based indexing
    # end_page remains as is because the range is exclusive at the end
    for page_num in range(start_page - 1, end_page):
        writer.add_page(reader.pages[page_num])

    with open(output_pdf, "wb") as out_f:
        writer.write(out_f)

# Usage
extract_pages("2025舞光LED21st(單頁水印可搜尋).pdf", "extracted_pages_10_14.pdf", 10, 14)
print("Done! Your new PDF is ready.")
```
